```python
#Dataset 2
student_grades = [["Suresh", 85],["Ramesh",92],["Rahul", 78],["Mohan", 78],["Raj", 60],["Arpit", 81],["Soham", 80]]
student_grades
```

```python
#1 Append New student Entry 
student_grades.append(["Sam",78])
student_grades
```

```python
#2 Insert a new student at some specific index
student_grades.insert (2,["Raman",99])
student_grades
```

```python
#3 Remove Specific Record 
student_grades.remove(["Arpit",81])
student_grades
```

```python
#4 Pop Terminal Record
removed = student_grades.pop(1)
removed
```

```python
student_grades
```

```python
#5 Count Duplicate marks of student
count = sum(1 for x in student_grades if x[1] == 78)
print(count)
```

```python
names_78 = [x[0] for x in student_grades if x[1] == 78]
print(names_78)
```

```python
#6 Merge with another Batch
student_grades.extend([["Shree",70] ,["Rocky", 75]])
student_grades
```

```python
#7 Reverse List Order
student_grades.reverse()
student_grades
```

```python
#8 Create Independent Backup (shallow copy)
backup = student_grades.copy()
backup.append(["NewStudent", 90])
backup
```

```python
student_grades
```

```python
#9 Update marks at Index 2
student_grades[2][1]=50
student_grades
```

```python
#10 Clear Database Records
student_grades.clear()
student_grades
```
